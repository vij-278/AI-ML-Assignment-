```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# YOUR RECORDED DATA
# -----------------------------
time = np.array([
0,30,60,90,120,150,180,210,240,270,
300,330,360,390,420,450,480,510,540,
570,600,630,660,690,720
])

velocity_kmph = np.array([
0,37,6,35,38,40,51,33,12,41,
43,38,39,31,35,39,40,45,42,
21,27,14,40,23,26
])

# Convert km/h → m/s
velocity_mps = velocity_kmph * (1000/3600)

# -----------------------------
# VELOCITY vs TIME
# -----------------------------
plt.figure()
plt.plot(time, velocity_kmph, marker='o')
plt.xlabel("Time (seconds)")
plt.ylabel("Velocity (km/h)")
plt.title("Velocity vs Time")
plt.grid()
plt.show()

# -----------------------------
# DISTANCE CALCULATION
# -----------------------------
distance = np.trapz(velocity_mps, time)

print(f"Calculated Distance = {distance:.2f} meters")

# -----------------------------
# DISTANCE vs TIME
# -----------------------------
distance_array = np.cumsum(velocity_mps * 30)  # 30 sec interval

plt.figure()
plt.plot(time, distance_array, marker='o')
plt.xlabel("Time (seconds)")
plt.ylabel("Distance (meters)")
plt.title("Distance vs Time")
plt.grid()
plt.show()

# -----------------------------
# ERROR CALCULATION
# -----------------------------
actual_distance = float(input("Enter actual distance from Google Maps (meters): "))

error = abs(actual_distance - distance)

print(f"Error = {error:.2f} meters")
print(f"Percentage Error = {(error/actual_distance)*100:.2f}%")
```
